__Exploring and Preparing Data__

We will first import the following libraries for this assignment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

__Exploratory Data Analysis__

We will first read the dataset and print out the summary.

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
df.head()
```

Let's plot the FlightNumber and PayloadMass variables to analyze how they impact the launch outcome. When the flight number increases, we can predict that the the first stage will come down successfully. If the payload is very massive, it is less likely the first stage will come back.

```python
sns.catplot(y = "PayloadMass", x = "FlightNumber", hue = "Class", data = df, aspect = 5)
plt.xlabel("Flight Number", fontsize = 25)
plt.ylabel("Pay load Mass (kg)", fontsize = 25)
plt.show()
```

We will evaluate some of the launch sites that are known for different success rates. CCAFS LC-40 has 60% of success while KSC LC-39A is at 77%.

__Step 1: Visualize the relationship between the Flight Number and the Launch Site__

The catplot function will help to create a plot of FlightNumber vs LaunchSite. To create the plot, set the parameter x to Flightnumber, y to Launch Site, and hue to 'class'.

```python
sns.catplot(y = "LaunchSite", x = "FlightNumber", hue = "Class", data = df, aspect = 5)
plt.xlabel("Flight Number", fontsize = 30)
plt.ylabel("LaunchSite", fontsize = 30)
plt.show()
```

__Step 2: Visualize the relationship between the Payload and the Launch Site__

Because we also want to get to know if there is a relationship between the launch sites and their payload mass, use the catplot method to create a plot for the topic "Launch Site vs. Payload Mass".

```python
sns.catplot(y = "LaunchSite", x = "PayloadMass", hue = "Class", data = df, aspect = 5)
plt.xlabel("Payload Mass", fontsize = 30)
plt.ylabel("Launch Site", fontsize = 30)
plt.show()
```

Most launch sites had a payload mass under 10,000 kg, but the exceptions were at CCAFS SLC 40 and KSC LC 39A, which implies they had the heavier mass.

__Step 3: Visualize the relationship between success rate of each orbit type__

Next, we want to check for any relationship between success rate and orbit type by creating a bar chart for the success rate of each orbit.

```python
df.groupby('Orbit')['Class'].mean().plot.bar()
```

__Step 4: Visualize the relationship between FlightNumber and Orbit type__

We will analyze the relationship between FlightNumber and Orbit type for each orbit if there exists any.

```python
sns.catplot(y = "Orbit", x = "FlightNumber", hue = "Class", data = df, aspect = 5)
plt.xlabel("Flight Number", fontsize = 30)
plt.ylabel("Orbit", fontsize = 30)
plt.show()
```

The success in the LEO orbit appears to be related to the number of flights, but there is no relationship between flight number and the GTO orbit.

__Step 5: Visualize the relationship between Payload and Orbit type__

Our next task is to plot the Payload vs Orbit scatter plot charts to show the relationship between the Payload and the Orbit type.

```python
sns.catplot(y = "Orbit", x = "PayloadMass", hue = "Class", data = df, aspect = 5)
plt.xlabel("Payload Mass", fontsize = 30)
plt.ylabel("Orbit", fontsize = 30)
plt.show()
```

Heavy payloads have both negative and positive influence on GTO orbits as well as positive influence on Polar LEO orbits.

__Step 6: Visualize the yearly trend of the launch success__

Let's plot a line chart with the x axis labeled as Year and y axis to be labeled as the average success rate to get the average launch success trend.

```python
year = []
def trend(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
trend(df['Date'])
```

```python
temp_df = df.copy()
temp_df['Year'] = year
temp_df.groupby('Year')['Class'].mean().plot()
```

The success rate increases from 2013 to 2020.
